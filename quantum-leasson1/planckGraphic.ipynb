```python
# After first lecture Victoria gave exercise to build the distribution of curves at different Temperature
# for Black Body radiation formula (Mathematica)
# planck_solution = B (_T, V_) :m (2 * h * v ^ 3) / c ^ 2 * 1 / (Exp [(h * v) / (kB * T)] - 1)

import matplotlib.pyplot as plt
import numpy as np

# 1. constants 
h = 6.626e-34   # Planck constant (J*s)
c = 3.0e8       # Speed of light (m/s)
kB = 1.38e-23   # Boltzmann constant (J/K)

# 2. planck function
def planck_frequency(nu, T):
    # term 1: 2hv^3 / c^2
    a = (2.0 * h * nu**3) / c**2
    
    # term 2: 1 / (exp(hv / (kB * T)) - 1)
    exponent = (h * nu) / (kB * T)
    b = 1.0 / (np.exp(exponent) - 1.0)

    return a * b

frequencies = np.linspace(1e12, 1.5e15, 500)
# set temperature
temperatures = [3000, 4000, 5000] # Kelvin


# 5. Plot 
plt.figure(figsize=(10, 6))

for T in temperatures:
    radiance = planck_frequency(frequencies, T)
    plt.plot(frequencies, radiance, label=f'T = {T} K')

planck_eq = r'$B_\nu(T) = \frac{2h\nu^3}{c^2} \frac{1}{e^{h\nu/k_B T} - 1}$'


# 6 add legends
plt.title(f'Planck Distribution\n{planck_eq}', fontsize=14)
plt.xlabel(r'Frequency $\nu$ (Hz)')
plt.ylabel(r'Spectral Radiance $B_\nu(T)$') 
plt.legend()
plt.grid(True, which="both", ls="-", alpha=0.5)
plt.tight_layout()
plt.show()
```
